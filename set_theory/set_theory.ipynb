# 1. Proving Identities
Given: A and B are sets, $\exists$ x such that (x $\in$ A $\land$ x $\in$ B)<br>
Show that:
- A\B = A $\cap$ $B^{c}$ <br>
Assume x $\in$ A \ B <br>
x $\in$ A and x $\notin$ B <br>
x $\in$ A and x $\in$ $B^c$ <br>
therefore x $\in$ A $\cap$ $B^c$ <br>
- B = (B $\cap$ A) $\cup$ (B $\cap$ $A^{c}$) <br>
assume x $\in$ (B $\cap$ A) $\cup$ (B $\cap$ $A^c$) <br>
x $\in$ B $\cap$ A or x $\in$ B $\cap$ $A^c$ <br>
if x $\in$ B $\cap$ A then x $\in$ B <br>
if x $\in$ B $\cap$ $A^c$ then x $\in$ B <br>
therefore x $\in$ B <br>
- B\A = B $\cap$ $A^{c}$ <br>
Assume x $\in$ B \ A <br>
x $\in$ B and x $\notin$ A <br>
x $\in$ B and x $\in$ $A^c$ <br>
therefore x $\in$ B $\cap$ $A^c$ <br>
- A $\cup$ B = A $\cup$ (B $\cap$ $A^{c}$) <br>
Assume x $\in$ A $\cup$ (B $\cap$ $A^c$) <br>
x $\in$ A or x $\in$ B $\cap$ $A^c$ <br>
if x $\in$ A then x $\in$ A $\cup$ B <br>
if x $\in$ B $\cap$ $A^c$ then x $\in$ A $\cup$ B <br>
therefore x $\in$ A $\cup$ B <br>
- $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ = $\cap_{i=1}^{\infty}A_{i}$ <br>
Apply De Morgan's law to the union of complements.
- |A $\cup$ B| = |A| + |B| - |A $\cap$ B| <br>
If set A and B are disjoint then |A $\cup$ B| = |A| + |B|, but if it's not then we count the elements of A and B then subtract it with |A $\cap$ B| once in order to avoid double-counting because the elements in the intersection are counted twice.

# 3. Set Operations

Given:<br>
$\textbf{U} = \{1,2,3,4,5\}$<br>
$\textbf{A} = \{1,2,3,4,5\}$<br>
$\textbf{B} = \{4,5,6,7\}$<br>
$\textbf{C} = \{5,6,7,8,9\}$<br>
$\textbf{D} = \{1,3,5,7,9\}$<br>
$\textbf{E} = \{2,4,6,8\}$<br>
$\textbf{F} = \{1,5,9\}$<br>
- $\textbf{C} \oplus \textbf{D}$<br>
C $\oplus$ D = (C $\cup$ D) \ (C $\cap$ D) = {1, 3, 6, 8}
- $\textbf{A} \backslash \textbf{B}$<br>
A \ B = {1, 2, 3}
- $\textbf{B} \backslash \textbf{A}$<br>
B \ A = {6, 7}
- $\textbf{E} \oplus \textbf{F}$<br>
E $\oplus$ F = (E $\cup$ F) \ (E $\cap$ F) = {1, 2, 4, 5, 6, 8, 9}

```python
U = {1, 2, 3, 4, 5}
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}
C = {5, 6, 7, 8, 9}
D = {1, 3, 5, 7, 9}
E = {2, 4, 6, 8}
F = {1, 5, 9}

print(C ^ D)
print(A - B)
print(B - A)
print(E ^ F)
```
